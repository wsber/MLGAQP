输出英文单词数 <= X 的元组后，将这些元组删除

```python
import re
import pandas as pd
from tqdm import tqdm

# 注册 tqdm 到 pandas
tqdm.pandas(desc="Processing rows")

# 读取完整 CSV
data_dir ='D:/softwareResource/datasets/IOGS/30W_valid_user/'
df = pd.read_csv(
    data_dir + 'post_LLM_cleaned_1.csv',
)

# 判断英文单词数是否 ≤ 4
def eng_word_count_le_2(text: str) -> bool:
    words = re.findall(r'[A-Za-z]+', str(text))
    return len(words) <= 4

# 1. 标记符合条件的行
mask = df['body'].progress_apply(eng_word_count_le_2)

# 2. 打印这些行
count = 0
for idx, row in df[mask].iterrows():
    print(f"id={row['id:ID']}\tbody={row['body']}\tML1_label={row['ML1_label']}\tLLM_label={row['LLM_label']}")
    count += 1
print(f"\n英文单词数 ≤ 4 的元组总数：{count}")

# 3. 删除这些行
df_filtered = df.loc[~mask].reset_index(drop=True)
print(f"删除后剩余行数：{len(df_filtered)}")

# 4. 保存过滤后的结果到新文件
output_path = data_dir + 'post_LLM_cleaned_2.csv'
df_filtered.to_csv(output_path, index=False)
print(f"已将删除后的数据保存到：{output_path}")
```
