```python
from transformers import pipeline

# 选择一个已经微调好的情感分析模型
# 这里以 CardiffNLP 推出的推特情感模型为例
sentiment = pipeline(
    "sentiment-analysis",
    model="cardiffnlp/twitter-roberta-base-sentiment",
    tokenizer="cardiffnlp/twitter-roberta-base-sentiment",
    return_all_scores=True,   # 如果你想同时拿到每个标签的分数
)

texts = [
    "I love this product! It's amazing 😀",
    "This is the worst movie I've ever seen..."
]

results = sentiment(texts)
for text, res in zip(texts, results):
    print(text)
    for label_score in res:
        print(f"  {label_score['label']}: {label_score['score']:.4f}")
    print()
```

统计/home/wangshuo/resource/datasets/parler/csv_data/pc/posts/valid_posts2 文件夹下post文件中有多少元组的body字段中的单词少于10个

```python
import os
import pandas as pd
from tqdm import tqdm

# 参数设定
posts_dir = '/home/wangshuo/resource/datasets/parler/csv_data/pc/posts/valid_posts2'
threshold = 3   # 单词数阈值

total_count = 0
per_file_counts = {}

for fname in tqdm(os.listdir(posts_dir), desc="Processing files"):
    if not fname.endswith('.csv'):
        continue
    path = os.path.join(posts_dir, fname)
    df = pd.read_csv(path, dtype=str)

    word_counts = df['body'].fillna('').str.split().apply(len)
    short_mask = word_counts < threshold
    count_short = short_mask.sum()

    per_file_counts[fname] = int(count_short)
    total_count += int(count_short)

    print(f"\n文件 `{fname}` 中共有 {count_short} 条记录，body 单词数 < {threshold}：")
    df_short = df.loc[short_mask, 'body']  # 只取 body 列

    if not df_short.empty:
        # 逐条输出 body 内容
        for i, text in enumerate(df_short, 1):
            print(f"{i:3d}. {text}")
    else:
        print("  （无此类记录）")

# 最终统计
print("\n—— 统计汇总 ——")
for fname, cnt in per_file_counts.items():
    print(f"  {fname:30s}: {cnt:4d}")
print(f"所有文件累计：{total_count} 条记录，body 单词数 < {threshold}")  
```
