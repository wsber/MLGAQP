输入的图名称为parler_0.graph文件， 按照顶点的标签降序排序，结果输出到parler_1.graph，并统计每个标签下有多少节点

```python
import collections

input_file = "/home/wangshuo/projects/FaSTest-main/dataset/parler_0/parler_0.graph"
output_file = "/home/wangshuo/projects/FaSTest-main/dataset/parler_0/parler_1.graph"

# 读取文件
with open(input_file, "r") as f:
    lines = f.readlines()

header = None
vertices = []
edges = []

for line in lines:
    if line.startswith("t"):
        header = line.strip().split()
    elif line.startswith("v"):
        parts = line.strip().split()
        vid = int(parts[1])
        label = int(parts[2])
        degree = int(parts[3])
        vertices.append((vid, label, degree))
    elif line.startswith("e"):
        parts = line.strip().split()
        src = int(parts[1])
        dst = int(parts[2])
        elabel = int(parts[3]) if len(parts) > 3 else 0
        edges.append((src, dst, elabel))

# 按 label 升序排序（保持 ID 不变）
vertices_sorted = sorted(vertices, key=lambda x: x[1])

# 统计每个标签的节点数量
label_count = collections.Counter([label for (_, label, _) in vertices_sorted])

print("标签统计（Label -> 节点数）:")
for label, count in sorted(label_count.items()):
    print(f"  {label}: {count}")

# 输出新文件
with open(output_file, "w") as f:
    f.write(f"t {len(vertices_sorted)} {len(edges)}\n")
    for vid, label, degree in vertices_sorted:
        f.write(f"v {vid} {label} {degree}\n")
    for src, dst, elabel in edges:
        if elabel != 0:
            f.write(f"e {src} {dst} {elabel}\n")
        else:
            f.write(f"e {src} {dst}\n")

print(f"\n✅ 已生成文件: {output_file}")
```

读取标签为x的所有节点，将这些节点的id保存到/home/wangshuo/projects/FaSTest-main/dataset/sv/parler_0/sv.txt文件夹下

```python
import os
from collections import defaultdict

# ==== 可修改参数 ====
input_file = "/home/wangshuo/projects/FaSTest-main/dataset/parler_0/parler_0.graph"
output_dir = "/home/wangshuo/projects/FaSTest-main/dataset/sv/svs"
# ====================

# === 1. 读取图文件 ===
label_to_nodes = defaultdict(list)

with open(input_file, "r") as f:
    for line in f:
        if line.startswith("v"):
            parts = line.strip().split()
            if len(parts) >= 3:
                vid = int(parts[1])
                label = int(parts[2])
                label_to_nodes[label].append(vid)

# === 2. 检查是否有节点 ===
if not label_to_nodes:
    print("[错误] 没有在图文件中找到任何顶点行（以 'v' 开头）。")
    exit()

# === 3. 确保输出文件夹存在 ===
os.makedirs(output_dir, exist_ok=True)

# === 4. 为每个标签单独保存文件 ===
for label, vids in sorted(label_to_nodes.items()):
    output_file = os.path.join(output_dir, f"sv_{label}.txt")
    with open(output_file, "w") as f:
        for vid in vids:
            f.write(f"{vid}\n")
    print(f"✅ 标签 {label}: {len(vids)} 个节点 -> {output_file}")

# === 5. 总结 ===
total_nodes = sum(len(v) for v in label_to_nodes.values())
print(f"\n🎯 共处理 {len(label_to_nodes)} 个标签，总节点数 {total_nodes}")
```

在 /home/wangshuo/projects/FaSTest-main/dataset/yeast/query_graph文件夹下搜索包含标签为x的查询，并将这些查询保存到文件夹/home/wangshuo/projects/FaSTest-main/dataset/partition_data/query_x_graph下，同时将/home/wangshuo/projects/FaSTest-main/dataset/yeast 下的 yeast_ans.txt 对应的真实答案写到 /home/wangshuo/projects/FaSTest-main/dataset/partition_data/的 ans.txt中 

```python
import os
import shutil

# === 参数设置 ===
LABEL = "6"  # 要搜索的节点标签（字符串类型）
query_dir = "/home/wangshuo/projects/FaSTest-main/dataset/yeast/query_graph"
ans_file = "/home/wangshuo/projects/FaSTest-main/dataset/yeast/yeast_ans.txt"
out_query_dir = f"/home/wangshuo/projects/FaSTest-main/dataset/partition_data/query_{LABEL}_graph"
out_ans_file = f"/home/wangshuo/projects/FaSTest-main/dataset/partition_data/query_{LABEL}_ans.txt"

os.makedirs(out_query_dir, exist_ok=True)

# === Step 1. 搜索包含指定标签的查询图 ===
selected_queries = set()

for fname in os.listdir(query_dir):
    if not fname.endswith(".graph"):
        continue
    fpath = os.path.join(query_dir, fname)
    try:
        contains_label = False
        with open(fpath, "r") as f:
            for line in f:
                line = line.strip()
                if not line or not line.startswith("v"):
                    continue
                parts = line.split()
                if len(parts) < 3:
                    continue
                vertex_label = parts[2]  # v [id] [label] [degree]
                if vertex_label == LABEL:
                    contains_label = True
                    break
        if contains_label:
            selected_queries.add(fname)
            shutil.copy(fpath, os.path.join(out_query_dir, fname))
    except Exception as e:
        print(f"[Warning] Failed to read {fname}: {e}")

print(f"[Info] 共选中了 {len(selected_queries)} 个包含标签 {LABEL} 的查询文件。")

# === Step 2. 过滤 yeast_ans.txt ===
num_ans_written = 0
with open(ans_file, "r") as fin, open(out_ans_file, "w") as fout:
    for line in fin:
        parts = line.strip().split()
        if len(parts) < 2:
            continue
        qname = parts[0]
        # 有的答案文件可能没有 .graph 后缀
        if qname.endswith(".graph") and qname in selected_queries:
            fout.write(line)
            num_ans_written += 1
        else:
            qfile = os.path.basename(qname)
            if not qfile.endswith(".graph"):
                qfile = qfile + ".graph"
            if qfile in selected_queries:
                fout.write(line)
                num_ans_written += 1

print(f"[Info] 已写入过滤后的真实答案文件：{out_ans_file}")
print(f"[Info] 共保留 {num_ans_written} 条答案记录。")
```

```python
import math
import re

# === 文件路径 ===
est_file = "/home/wangshuo/projects/FaSTest-main/dataset/sv/sv_estimate_result.txt"
ans_file = "/home/wangshuo/projects/FaSTest-main/dataset/yeast_exp/yeast_exp_ans.txt"

# === 读取真实值 ===
true_values = {}
with open(ans_file, "r") as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) >= 3:
            query_name = parts[0].strip()
            try:
                true_val = float(parts[-1])
                true_values[query_name] = true_val
            except ValueError:
                pass

# === 读取估计结果 ===
results = {}
current_query = None
current_sum = 0.0
current_all_est = None

with open(est_file, "r") as f:
    for line in f:
        line = line.strip()
        if line.startswith("Query:"):
            # 保存上一个 query 的结果
            if current_query is not None:
                results[current_query] = (current_sum, current_all_est)
            # 提取新 query 名称
            m = re.search(r"([^/\\]+\.graph)", line)
            if m:
                current_query = m.group(1)
                current_sum = 0.0
                current_all_est = None

        elif line.startswith("All Est:"):
            try:
                current_all_est = float(line.split(":")[1].strip())
            except ValueError:
                current_all_est = None

        elif line.startswith("----------------------------------------"):
            if current_query is not None:
                results[current_query] = (current_sum, current_all_est)
                current_query = None

        else:
            # 节点估计行
            parts = line.split()
            if len(parts) >= 2:
                try:
                    val = float(parts[1])
                    current_sum += val
                except ValueError:
                    pass

# 最后一段
if current_query is not None:
    results[current_query] = (current_sum, current_all_est)

# === 计算 log(Q_error) 并收集排序用数据 ===
records = []
for q, (est_sum, all_est) in results.items():
    true_val = true_values.get(q)
    if true_val is None or true_val <= 0:
        log_q_error = float("inf")
        log_q_error_all = float("inf")
    else:
        # 对 sum 与 all_est 分别计算误差
        log_q_error = (
            float("inf")
            if est_sum <= 0
            else math.log10(max(est_sum / true_val, true_val / est_sum))
        )
        log_q_error_all = (
            float("inf")
            if all_est is None or all_est <= 0
            else math.log10(max(all_est / true_val, true_val / all_est))
        )
    records.append((q, est_sum, all_est, true_val, log_q_error, log_q_error_all))

# === 按 log(Q_error_AllEst) 降序排列 ===
records.sort(key=lambda x: (float("inf") if math.isnan(x[5]) else x[5]), reverse=True)

# === 输出结果 ===
print(f"{'Query':40} {'SumEst':>12} {'AllEst':>12} {'TrueVal':>12} {'logQ(Sum)':>12} {'logQ(All)':>12}")
print("-" * 95)
for q, est_sum, all_est, true_val, log_q_error, log_q_error_all in records:
    tval = "N/A" if true_val is None else f"{true_val:.0f}"
    log_sum = "INF" if math.isinf(log_q_error) else f"{log_q_error:.6f}"
    log_all = "INF" if math.isinf(log_q_error_all) else f"{log_q_error_all:.6f}"
    all_est_str = "N/A" if all_est is None else f"{all_est:.8f}"
    print(f"{q:40} {est_sum:12.8f} {all_est_str:>12} {tval:>12} {log_sum:>12} {log_all:>12}")
```
